```python
"""CIS2532 NET 01
Konstantinos Spyropoulos
HW 3 Question 2
02/19/22
Program: Using pandas to calculate information from the Titanic.csv"""

import csv
import pandas as pd

# read data from Titanic.csv
data = pd.read_csv('C:\\Users\\kspyr\\Desktop\\CIS 2532\\CIS2532 HW 3\\titanic.csv')

# put data from Titanic.csv in a dataframe
df = pd.DataFrame(data)

# counts amount of passengers in Titanic.csv
count = df.index
totalPassengers = len(count)

print('part a')

# print total passengers in Titanic
print("There were "+str(totalPassengers)+" passengers in the Titanic")

print()

print('part b')

# set males variable to zero and use it as a counter to count male passengers on Titanic
males = 0

# for loop calculates male passengers on Titanic
for male in df.loc[df['sex'] == 'male', "Name"]:
    males = males + 1

# set females variable to zero and use it as a counter to count female passengers on Titanic
females = 0

# for loop calculates female passengers on Titanic
for female in df.loc[df['sex'] == 'female', "Name"]:
    females = females + 1

# printe male and female passenger amounts on Titanic
print("There are "+str(males)+" males")
print("There are "+str(females)+" females")

print()

print('part c')

# df['age'].mean() in dataframes calculates average of the age column in the csv file
# which calculates average age of passengers who boarded the Titanic
print('Average age of all passengers:'+str(df['age'].mean()))

print()

print('part d')

# over21Group represents the collection of passenger ages over 21
over21Group = df.loc[(df["age"] < 21),"age"]

# counter variable is set to zero and is used to count how many passengers
# are over 21 who boarded the Titanic
peopleOver21 = 0 

# for loop calculates amount of passengers over 21
for person in over21Group:
    peopleOver21 = peopleOver21 + 1

# print amount of passengers who boarded the Titanic who were under 21
print("Amount of passengers under 21: " + str(peopleOver21))


print()

print('part e')

# maleSurvivedGroup stores the collection of male names who survived the Titanic
malesSurvivedGroup = df.loc[((df["survived"] == "yes") & (df["sex"] == "male")),"Name"]

# malesSurvived is counter set to 0 that will count total amount of male survivors
malesSurvived = 0

# for loop calculates amount of male survivors
for person in malesSurvivedGroup:
    malesSurvived = malesSurvived + 1
    
# femaleSurvivedGroup stores the collection of female names who survived the Titanic
femalesSurvivedGroup = df.loc[((df["survived"] == "yes") & (df["sex"] == "female")),"Name"]

# femaleSurived is counter set to 0 that will count total amount of female survivors
femalesSurvived = 0

# for loop calculates amount of female survivors
for person in femalesSurvivedGroup:
    femalesSurvived = femalesSurvived + 1

# malesDeadGroup stores the collection of male names who died on the Titanic
malesDeadGroup = df.loc[((df["survived"] == "no") & (df["sex"] == "male")),"Name"]

# malesDead is counter set to 0 that will count total amount of dead males 
malesDead = 0

# for loop calculates amount of dead males
for person in malesDeadGroup:
    malesDead = malesDead + 1

# femalesDeadGroup stores the collection of female names who died on the Titanic
femalesDeadGroup = df.loc[((df["survived"] == "no") & (df["sex"] == "female")),"Name"]

# femalesDead is counter set to 0 that will count total amount of dead females 
femalesDead = 0

# for loop calculates amount of dead males
for person in femalesDeadGroup:
    femalesDead = femalesDead + 1

print("Total Passengers Survived: "+ str(malesSurvived + femalesSurvived)+",Male: "+str(malesSurvived)+",Female: "+str(femalesSurvived))
print("Total Passengers Dead: "+ str(malesDead + femalesDead)+",Male: "+str(malesDead)+",Female: "+str(femalesDead))

print()

print('part f')

# nameOldestSurvivor finds name of oldest survivor on the Titanic and returns a view object using .values
nameOldestSurvivor = df.loc[df['age'] == df['age'].max(),"Name"].values

# nameOldestSurvivorShown shows only the name of oldest survivor with no brackets
nameOldestSurvivorShown =(','.join(nameOldestSurvivor))

# prints name of oldest survivor
print("Name of oldest survivor: "+str(nameOldestSurvivorShown))

# ageOldestSurvivor finds age of oldest survivor on the Titanic and returns a view object using .values
ageOldestSurvivor = str(df.loc[df['age'] == df['age'].max(),"age"].values)

# ageOldestSurvivorShown gets rid of any brackets or commas and just shows the age of the oldest survivor
ageOldestSurvivorShown = str(ageOldestSurvivor)[1:-1]

# print age of oldest survivor
print("Age of oldest survivor: "+str(ageOldestSurvivorShown))

# nameYoungestSurvivor finds name of youngest survivor on the Titanic and returns a view object using .values
nameYoungestSurvivor = df.loc[df['age'] == df['age'].min(),"Name"].values

# nameYoungestSurvivorShown shows only the name of oldest survivor with no brackets
nameYoungestSurvivorShown = (','.join(nameYoungestSurvivor))

# print name of youngest survivor
print("Name of youngest survivor: "+str(nameYoungestSurvivorShown))

# ageYoungestSurvivor finds age of youngest survivor on the Titanic and returns a view object using .values
ageYoungestSurvivor = str(df.loc[df['age'] == df['age'].min(),"age"].values)

# ageYoungestSurvivorShown gets rid of any brackets or commas and just shows the age of the youngest survivor
ageYoungestSurvivorShown = str(ageYoungestSurvivor)[1:-1]

# print age of youngest survivor
print("Age of youngest survivor: "+str(ageYoungestSurvivorShown))

print()

print('part g')
print()
print('List of passengers')

# passengerList contains a collection of names of those who survived the Titanic
passengerList = df.loc[df["survived"] == "yes","Name"]

# for loop prints the name of every survivor of the Titanic
for person in passengerList:
    print(person)


```
